2. Write a function that will plot the ranking of a given roller coaster over time as a line. Your function should take a roller coaster's name and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with `"El Toro"` as the roller coaster name and the wood ranking DataFrame. What issue do you notice? Update your function with an additional argument to alleviate the problem, and retest your function.

```python
### 2
# Create a function to plot rankings over time for 1 roller coaster
def roller_rank(name, type, park):
    if type == 'wood':
        wood_roller = wood[(wood.Name == name) & (wood.Park == park)]
        plt.plot(wood_roller['Year of Rank'], wood_roller['Rank'], label=name)
        plt.xlabel('Ranking Year')
        plt.ylabel('Ranking')
        plt.title('Roller Coaster Ranking per Year at {}'.format(park))
        plt.legend()
        plt.show()
        return wood_roller
    elif type == 'steel':
        steel_roller = steel[(steel.Name == name) & (steel.Park == park)]
        plt.plot(steel_roller['Year of Rank'], steel_roller['Rank'], label=name)
        plt.xlabel('Ranking Year')
        plt.ylabel('Ranking')
        plt.title('Roller Coaster Ranking per Year at {}'.format(park))
        plt.legend()
        plt.show()
        return steel_roller
    else:
        print('Only specify wood or steel roller coasters.')

    

# Create a plot of El Toro ranking over time
roller_rank('El Toro', 'wood', 'Six Flags Great Adventure')
```

3. Write a function that will plot the ranking of two given roller coasters over time as lines. Your function should take both roller coasters' names and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with `"El Toro"` as one roller coaster name, `"Boulder Dash"` as the other roller coaster name, and the wood ranking DataFrame. What issue do you notice? Update your function with two additional arguments to alleviate the problem, and retest your function.

```python
# 3
# Create a function to plot rankings over time for 2 roller coasters
def two_roller_rank(name1, name2, type, park1, park2):
    ax = plt.subplot()
    if type == 'wood':
        wood_roller_1 = wood[(wood.Name == name1) & (wood.Park == park1)]
        wood_roller_2 = wood[(wood.Name == name2) & (wood.Park == park2)]
        ax.plot(wood_roller_1['Year of Rank'], wood_roller_1['Rank'], label=name1)
        ax.plot(wood_roller_2['Year of Rank'], wood_roller_2['Rank'], label=name2)
        ax.invert_yaxis()
        plt.xlabel('Ranking Year')
        plt.ylabel('Ranking')
        plt.title('Roller Coaster Ranking per Year')
        plt.legend()
        plt.show()
        return wood.Park[(wood.Name == name1) | (wood.Name == name2)].unique()
    elif type == 'steel':
        steel_roller_1 = steel[(steel.Name == name1) & (steel.Park == park1)]
        steel_roller_2 = steel[(steel.Name == name2) & (steel.Park == park2)]
        ax.plot(steel_roller_1['Year of Rank'], steel_roller_1['Rank'], label=name1)
        ax.plot(steel_roller_2['Year of Rank'], steel_roller_2['Rank'], label=name2)
        plt.xlabel('Ranking Year')
        plt.ylabel('Ranking')
        plt.title('Roller Coaster Ranking per Year')
        plt.legend()
        plt.show()
        return steel.Park[(steel.Name == name1) | (steel.Name == name2)].unique()
    else:
        print('Only specify wood or steel roller coasters.')
# Create a plot of El Toro and Boulder Dash roller coasters
two_roller_rank('El Toro', 'Boulder Dash', 'wood', 'Six Flags Great Adventure', 'Lake Compounce')
```

```python
# 4
# Create a function to plot top n rankings over time
def top_roller_rank(type, top):
    if type == 'wood':
        top_wood = wood[wood.Rank <= top]
        fig, ax = plt.subplots(figsize=(10,10))
        for name in top_wood.Name.unique():
            data = top_wood[top_wood.Name == name]
            ax.plot(data['Year of Rank'], data['Rank'], label=name)
        plt.xlabel('Ranking Year')
        plt.ylabel('Ranking')
        plt.title('Top Roller Coaster Rankings')
        ax.set_yticks([i for i in range(1, (top+1))])
        ax.invert_yaxis()
        plt.legend(loc=4)
        plt.show()        
        
    elif type == 'steel':
        top_steel = steel[steel.Rank <= top]
        fig, ax = plt.subplots(figsize=(10,10))
        for name in top_steel.Name.unique():
            data = top_steel[top_steel.Name == name]
            ax.plot(data['Year of Rank'], data['Rank'], label=name)
        plt.xlabel('Ranking Year')
        plt.ylabel('Ranking')
        plt.title('Top Roller Coaster Rankings')
        ax.set_yticks([i for i in range(1, (top+1))])
        ax.invert_yaxis()
        plt.legend(loc=0)
        plt.show()  
    else:
        print('Only specify wood or steel roller coasters.')  

# Create a plot of top n rankings over time
top_roller_rank('wood', 5)
```

7. Write a function that creates a bar chart showing the number of inversions for each roller coaster at an amusement park. Your function should take the roller coaster DataFrame and an amusement park name as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame and amusement park name.

```python
# 7
# Create a function to plot inversions by coaster at park
roller_coaster = pd.read_csv(r"C:\Users\jugil\Documents\Cursos\CodeAcademy\Data Analytics Career Path\3. Statistics for Data Science\Visualization for Data Science Applications\Roller Coaster Project\roller_coasters.csv")

def inv_coaster(park, data):
    inv_park_coaster = data.groupby(['park', 'name']).num_inversions.sum().reset_index()
    inv_park_coaster_filt = inv_park_coaster[inv_park_coaster.park == park].sort_values('num_inversions', ascending=False)
    plt.bar(range(len(inv_park_coaster_filt)), inv_park_coaster_filt.num_inversions)
    ax = plt.subplot()
    ax.set_xticks(range(len(inv_park_coaster_filt)))
    ax.set_xticklabels(inv_park_coaster_filt.name)
    plt.xlabel('Roller Coaster Name')
    plt.ylabel('Number of Inversions')
    plt.title('Total Number of Inversions per Roller Coaster')
    plt.show()
    return inv_park_coaster[inv_park_coaster.park == park]


# Create barplot of inversions by roller coasters
inv_coaster('Adlabs Imagica', roller_coaster)
```

8. Write a function that creates a pie chart that compares the number of operating roller coasters (`'status.operating'`) to the number of closed roller coasters (`'status.closed.definitely'`). Your function should take the roller coaster DataFrame as an argument. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame.

```python
# 8
# Create a function to plot a pie chart of status.operating
def status(data):
    status_count = roller_coaster.status.value_counts().reset_index()
    status_count = status_count.rename({'count':'status_type'}, axis=1)
    status_count = status_count[(status_count.status == 'status.operating') | (status_count.status == 'status.closed.definitely')]
    plt.pie(status_count.status_type, labels=['Operating', 'Not Operating'], autopct='%d%%')
    plt.axis('equal')
    plt.title('Operating vs Not Operating Roller Coasters')
    plt.show()
    
#Create pie chart of roller coasters
status(roller_coaster)
```

9. `.scatter()` is another useful function in matplotlib that you might not have seen before. `.scatter()` produces a scatter plot, which is similar to `.plot()` in that it plots points on a figure. `.scatter()`, however, does not connect the points with a line. This allows you to analyze the relationship between two variables. Find [`.scatter()`'s documentation here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).
